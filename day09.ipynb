## Part 1

```python
with open("input/day9.txt") as f:
    data = f.readlines()
```

```python
data = [list(row.strip()) for row in data]
data = [[int(num) for num in row] for row in data]
```

```python
low_points = []

for i in range(len(data)):
    for j in range(len(data)):
        
        point = data[i][j]
        adj_points = []
        
        # UP
        if j-1<0:
            pass
        else:
            adj_points.append(data[i][j-1])
            
        # DOWN
        if j+1>=len(data):
            pass
        else:
            adj_points.append(data[i][j+1])
            
        # LEFT
        if i-1<0:
            pass
        else:
            adj_points.append(data[i-1][j])
            
        # RIGHT
        if i+1>=len(data):
            pass
        else:
            adj_points.append(data[i+1][j])
        
        if all(ap > point for ap in adj_points):
            low_points.append(point)
```

```python
result = len(low_points) + sum(low_points)
```

## Part 2

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

from scipy.ndimage import measurements
from collections import Counter
from functools import reduce
```

```python
data = [[int(num) for num in row] for row in data]
```

```python
data_binary = [[1 if num < 9 else 0 for num in row] for row in data]
```

```python
basins, num_basins = measurements.label(data_binary)
```

```python
f = figure(figsize=(20, 20))

f.add_subplot(1, 3, 1)
plt.imshow(data, cmap="Blues")
plt.title("Original")

f.add_subplot(1, 3, 2)
plt.imshow(data_binary, cmap="Blues")
plt.title("Binary")

f.add_subplot(1, 3, 3)
plt.imshow(basins, cmap="Blues")
plt.title("Clustered");
```

```python
biggest_basins = Counter(sum([list(row) for row in list(basins)],[])).most_common()[1:4]
result = reduce(lambda x,y: x*y , [v for k,v in biggest_basins])
```
